### Imputing Missing 'Age' Values
Missing values in the 'Age' column are filled with the mean age of all passengers to maintain data integrity.

```python
df['Age'].fillna(df['Age'].mean(), inplace=True)
```

### Imputing Missing 'Cabin' Values
Missing values in the 'Cabin' column are filled with the most frequently occurring cabin value.

```python
df['Cabin'].fillna(df['Cabin'].mode()[0], inplace=True)
```

### Imputing Missing 'Embarked' Values
Missing values in the 'Embarked' column are filled with the most frequently occurring embarkation port.

```python
df['Embarked'].fillna(df['Embarked'].mode()[0], inplace=True)
```

### Verifying No More Missing Values
After imputation, this cell re-checks for any remaining missing values across all columns.

```python
df.isna().sum()
```

### Visualizing Male vs. Female Survival
This pie chart illustrates the proportion of male and female passengers who survived, including their respective counts and percentages.

```python
### Now it's time to visuallie the data throug Pandas Graphs

males_servived = df[(df['Sex'] == 'male') & (df['Survived'] == 1)].shape[0]
females_survived = df[(df['Sex'] == 'female') & (df['Survived'] == 1)].shape[0]

# Now, plotting the pie graph of males and females survived
import matplotlib.pyplot as plt

labels = ['Males Survived', 'Females Survived']
sizes = [males_servived, females_survived]
colors = ['lightskyblue', 'lightcoral']
explode = (0.1, 0)  # explode 1st slice (Females Survived)

plt.figure(figsize=(8, 8))
plt.pie(sizes, explode=explode, labels=labels, colors=colors, autopct='%1.1f%%',
        shadow=True, startangle=140)
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.title('Percentage of Male and Female Survivors')
plt.show()
```

### Analyzing Survival Rate by Age
This line graph shows how the survival rate varies across different age groups, providing insights into age-related survival trends.

```python
## Now I want to plot the line graph of Age and Survived, on x=Age, y=Survived

import matplotlib.pyplot as plt
import seaborn as sns

# Calculate the mean of 'Survived' for each 'Age'
survival_rate_by_age = df.groupby('Age')['Survived'].mean().reset_index()

plt.figure(figsize=(12, 6))
sns.lineplot(data=survival_rate_by_age, x='Age', y='Survived')
plt.title('Survival Rate by Age')
plt.xlabel('Age')
plt.ylabel('Survival Rate')
plt.grid(True)
plt.show()
```

### Exploring Survival Rate by Passenger Class
This bar plot visualizes the survival rate for each passenger class, highlighting potential differences in survival chances based on class.

```python
## Plotting the Survival Rate by Class
import matplotlib.pyplot as plt
import seaborn as sns

# Calculate the mean of 'Survived' for each 'Pclass'
survival_rate_by_class = df.groupby('Pclass')['Survived'].mean().reset_index()

plt.figure(figsize=(8, 6))
sns.barplot(data=survival_rate_by_class, x='Pclass', y='Survived', palette='viridis')
plt.title('Survival Rate by Passenger Class')
plt.xlabel('Passenger Class')
plt.ylabel('Survival Rate')
plt.ylim(0, 1) # Survival rate is between 0 and 1
plt.grid(axis='y')
plt.show()
```
